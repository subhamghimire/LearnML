## Training and Test Dataset Split

```python
features = df.drop('PRICE',axis=1)
prices = df['PRICE']

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.33, random_state=42)

# % of training set
len(X_train)/len(features)
```

```python
# % of test data set
# len(X_test)/len(features)
X_test.shape[0]/features.shape[0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()

reg.fit(X_train, y_train)
```

```python
print('Intercept', reg.intercept_)
pd.DataFrame(data=reg.coef_, index=X_train.columns, columns=['coef'])
```

```python
print('Training data r-squared ', reg.score(X_train, y_train))
print('Test data r-squared ', reg.score(X_test, y_test))
```

## Regression using log prices

```python
features = df.drop('PRICE',axis=1)
prices = np.log(df['PRICE']) #use log prices

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.33, random_state=42)

reg = LinearRegression()
reg.fit(X_train, y_train)

print('Training data r-squared ', reg.score(X_train, y_train))
print('Test data r-squared ', reg.score(X_test, y_test))

print('Intercept', reg.intercept_)
pd.DataFrame(data=reg.coef_, index=X_train.columns, columns=['coef'])
```

```python
#Charles River Property Premium
np.e**(0.106790)
```

## p values and evaluating coefficients

```python
import statsmodels.api as sm
```

```python
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# results.params
# results.pvalues

pd.DataFrame({'coef':results.params,'p-value': round(results.pvalues,3)})
```

## Testing for multicollinearity

$$ TAX = \alpha _0 + \alpha _1 RM + \alpha _2 NOX + ...+ \alpha _{12}LSTAT      $$
$$ VIF _{TAX} = \frac{1}{(1- R_{TAX}^ 2)} $$

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variance_inflation_factor(exog=X_incl_const.values, exog_idx=1)
# type(X_incl_const)
```

```python
# len(X_incl_const.columns)
# X_incl_const.shape[1]
```

```python
vif = []
for i in range(X_incl_const.shape[1]):
    vif.append(variance_inflation_factor(exog=X_incl_const.values, exog_idx=i))
    
print(vif)
```

```python
vif = [variance_inflation_factor(exog=X_incl_const.values, exog_idx=i) for i in range(X_incl_const.shape[1])]
    
pd.DataFrame({'coef_name':X_incl_const.columns,'vif': np.around(vif,2)})
```

## Model Simplicity & the BIC (Bayesian information criterion)

```python
#Original model with log prices and all features
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()


org_coef = pd.DataFrame({'coef':results.params,'p-value': round(results.pvalues,3)})

print('BIC is: ',results.bic)
print('r-squared is: ',results.rsquared)
```

```python
#Reduced model #exclusing INDUS
X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'],axis=1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()


coef_minus_indus = pd.DataFrame({'coef':results.params,'p-value': round(results.pvalues,3)})

print('BIC is: ',results.bic)
print('r-squared is: ',results.rsquared)
```

```python
#Reduced model #exclusing INDUS and AGE
X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS','AGE'],axis=1)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()


reduced_coef = pd.DataFrame({'coef':results.params,'p-value': round(results.pvalues,3)})

print('BIC is: ',results.bic)
print('r-squared is: ',results.rsquared)
```

```python
frames = [org_coef, coef_minus_indus, reduced_coef]
pd.concat(frames, axis=1)
```

## Residuals and Residual Plots

```python
# Modified model: transformed & simplified (drop 2 features)
features = df.drop(['PRICE','INDUS', 'AGE'],axis=1)
prices = np.log(df['PRICE']) #use log prices

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.33, random_state=42)

#using statsmodel
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()

#residuals
# residuals = y_train - results.fittedvalues
# residuals
# results.resid

# Graph of Actual vs. Predicted Prices
corr = round(y_train.corr(results.fittedvalues),2)
plt.scatter(x=y_train,y=results.fittedvalues, c='navy', alpha=0.5)
plt.plot(y_train, y_train, color='cyan')
plt.title(f'Actual vs Predicted log prices: $y _i$ vs $\hat y _i$(Corr={corr})')
plt.xlabel('Actual log prices $y _i$', fontsize=14)
plt.ylabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.show()


plt.scatter(x=np.e**y_train,y=np.e**results.fittedvalues, c='blue', alpha=0.5)
plt.plot(np.e**y_train,np.e**y_train, color='cyan')
plt.title(f'Actual vs Predicted prices: $y _i$ vs $\hat y _i$(Corr={corr})')
plt.xlabel('Actual prices $y _i$', fontsize=14)
plt.ylabel('Predicted prices $\hat y _i$', fontsize=14)
plt.show()

# Residual vs predicted values
plt.scatter(x=results.fittedvalues,y=results.resid, c='blue', alpha=0.5)

plt.title('Residuals vs Fitted values')
plt.xlabel('Prediced log prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.show()


# Mean squared error & r-aquared
reduced_log_mse = round(results.mse_resid, 3)
reduced_log_rsquared = round(results.rsquared, 3)
```

```python
# Distrubution of Residuals (log prices) - checking for normality
resid_mean = round(results.resid.mean(),3)
resid_skew = round(results.resid.skew(),3)

sns.distplot(results.resid,color='navy')
plt.title(f'Log price model: residuals Skew ({resid_skew}) Mean({resid_mean})')
plt.show()
```

```python
# Original model: normal prices and all features
features = df.drop(['PRICE'],axis=1)
prices = df['PRICE'] 

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.33, random_state=42)

#using statsmodel
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()


# Graph of Actual vs. Predicted Prices
corr = round(y_train.corr(results.fittedvalues),2)
plt.scatter(x=y_train,y=results.fittedvalues, c='indigo', alpha=0.5)
plt.plot(y_train, y_train, color='cyan')
plt.title(f'Actual vs Predicted  prices: $y _i$ vs $\hat y _i$(Corr={corr})')
plt.xlabel('Actual prices $y _i$', fontsize=14)
plt.ylabel('Predicted prices $\hat y _i$', fontsize=14)
plt.show()


# Residual vs predicted values
plt.scatter(x=results.fittedvalues,y=results.resid, c='indigo', alpha=0.5)

plt.title('Residuals vs Fitted values')
plt.xlabel('Prediced log prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.show()

#Residual distrubution chart
resid_mean = round(results.resid.mean(),3)
resid_skew = round(results.resid.skew(),3)

sns.distplot(results.resid,color='indigo')
plt.title(f'Residuals Skew ({resid_skew}) Mean({resid_mean})')
plt.show()

# Mean squared error & r-aquared
full_normal_mse = round(results.mse_resid, 3)
full_normal_rsquared = round(results.rsquared, 3)
```

```python
# Model Omitting Key features using log prices
features = df.drop(['PRICE','INDUS','AGE','LSTAT','RM','NOX','CRIM'],axis=1)
prices = np.log(df['PRICE'] )

X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.33, random_state=42)

#using statsmodel
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train, X_incl_const)
results = model.fit()


corr = round(y_train.corr(results.fittedvalues),2)
plt.scatter(x=y_train,y=results.fittedvalues, c='blue', alpha=0.5)
plt.plot(y_train, y_train, color='cyan')
plt.title(f'Actual vs Predicted  prices with omitted variables $y _i$ vs $\hat y _i$(Corr={corr})')
plt.xlabel('Actual log prices $y _i$', fontsize=14)
plt.ylabel('Predicted log prices $\hat y _i$', fontsize=14)
plt.show()


# Residual vs predicted values
plt.scatter(x=results.fittedvalues,y=results.resid, c='blue', alpha=0.5)

plt.title('Residuals vs Fitted values')
plt.xlabel('Prediced prices $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.show()


# Mean squared error & r-aquared
omitted_var_mse = round(results.mse_resid, 3)
omitted_var_rsquared = round(results.rsquared, 3)
```

```python
pd.DataFrame({'R-Squared':[reduced_log_rsquared, full_normal_rsquared, omitted_var_rsquared],
             'MSE':[reduced_log_mse, full_normal_mse, omitted_var_mse],
             'RMSE': np.sqrt([reduced_log_mse, full_normal_mse, omitted_var_mse])},
             index=['Reduced Log Model', 'Full Normal Model','Omitted Var Model'])
```

```python
print('1 s.d in log prices',np.sqrt(reduced_log_mse))
print('2 s.d in log prices',np.sqrt(reduced_log_mse)*2)

upper_bound = np.log(30) + np.sqrt(reduced_log_mse)*2
print('The upper bound for a 95% pred. in log prices  interval is',upper_bound)
print('The upper bound in normal price is',np.e**upper_bound * 1000)

lower_bound = np.log(30) - np.sqrt(reduced_log_mse)*2
print('The lower bound for a 95% pred. in log prices interval is',lower_bound)
print('The lower bound in normal price is',np.e**lower_bound * 1000)
```
